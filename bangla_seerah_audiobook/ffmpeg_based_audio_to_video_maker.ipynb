```python
%%time
# a big thanks to https://superuser.com/questions/1041816/combine-one-image-one-audio-file-to-make-one-video-using-ffmpeg
path = "videobook_translated_bangla_seerah"
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)
    
retry = [9,20,34,44,51,55,69,73,74,84,86,89]


for i in range(len(files)):
    if(i+1 not in retry):
        continue
    
    fpath = folder_path+'/'+files[i] 
    
    head, output = os.path.split(files[i])
    
    fname = output.split(".")[0]
    fname = fname.replace('\'','')
    
    fpath_new = re.sub(r'[\\*?:()&"<>|]',"_",fpath)
    os.rename(fpath, fpath_new)
    fpath_new = fpath_new.replace('\'','')
    print("working on ",fpath_new)
    fname = re.sub(r'[\\*?:()&"<>|]',"_",fname)
    out_fname = f'./{path}/{fname}.mp4'
    cmd = f'ffmpeg -loop 1 -i ./logoCombined2.png -i {fpath_new} -c:v libx264 -tune stillimage -c:a aac -b:a 192k -pix_fmt yuv420p -shortest {out_fname}'
    os.system(cmd)
```
