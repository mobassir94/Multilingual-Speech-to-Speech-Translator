```python

def mlt_reader(df):
#     bn_audio = []
#     curr_chapter = '১'
    
    for i in range(len(df)):
        book = df.iloc[i]['title'].replace(' ','_')
        mlt_text = df.iloc[i]['seerah_bn']
        print("index = ",i," Title = ",book)
        text = re.sub('পিবিইউএইচ', 'সাল্লাল্লাহু আলাইহি ওয়া সাল্লাম', mlt_text)
        text = re.sub('পিবিইউ', 'সাল্লাল্লাহু আলাইহি ওয়া সাল্লাম', text)
        text = re.sub('পিবিউ', 'সাল্লাল্লাহু আলাইহি ওয়া সাল্লাম', text)
        text = re.sub('পি.বি.ইউ.এইচ', 'সাল্লাল্লাহু আলাইহি ওয়া সাল্লাম', text)
        text = re.sub('এসডব্লিউটি', 'সুবহানাহুওয়াতালা', text)
        text = text.replace('(সা)', 'সাল্লাল্লাহু আলাইহি ওয়া সাল্লাম')
        text = text.replace('(আ)', 'আলাইহিস সালাম')  
        text = re.sub('এএস', 'আলাইহিস সালাম', text) 
        mlt_text = re.sub('সিরাহ্র', 'সীরাহ', text)

        audio=MLT_TTS(mlt_text,resample_audio_to_out_sample_rate=True)
#         bn_audio.append(audio)
#         full_chapter_audio=np.concatenate(bn_audio,axis=-1)
        sf.write(f'./{folder}/{book}.wav', audio, 22050)



```
